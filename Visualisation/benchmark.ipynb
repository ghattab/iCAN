```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import matplotlib.patches as mpatches
import os
```

```python
path_benchmark = os.path.join('..', 'Results', 'Benchmark')

df1 = pd.read_csv(os.path.join(path_benchmark, '1_benchmark_runs_5.csv'), index_col=0)
df2 = pd.read_csv(os.path.join(path_benchmark, '2_benchmark_runs_5.csv'), index_col=0)
df3 = pd.read_csv(os.path.join(path_benchmark, '3_benchmark_runs_5.csv'), index_col=0)
df4 = pd.read_csv(os.path.join(path_benchmark, '4_benchmark_runs_5.csv'), index_col=0)
df5 = pd.read_csv(os.path.join(path_benchmark, '5_benchmark_runs_5.csv'), index_col=0)
df6 = pd.read_csv(os.path.join(path_benchmark, '6_benchmark_runs_5.csv'), index_col=0)
df7 = pd.read_csv(os.path.join(path_benchmark, '7_benchmark_runs_5.csv'), index_col=0)
```

```python
absolute_times_df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
```

```python
RKI_colors = ['#005EB8','#71BAFE', '#F5F5F5', '#FECA76', '#B87200']

fig, ax = plt.subplots(figsize=(10,10))
plt.xscale('log')
plt.yscale('log')
plt.grid(axis = 'both')
plt.scatter(x=absolute_times_df['original_file_size'], y=absolute_times_df['level_1_data_driven'], c=RKI_colors[1])
plt.scatter(x=absolute_times_df['original_file_size'], y=absolute_times_df['level_2_data_driven'], c=RKI_colors[0])
ax.set_xlabel('Size of FASTA file in Bytes', fontsize='large')
ax.set_ylabel('Encoding time in s', fontsize='large')
ax.legend(['Encoding first-level neighbourhoods', 'Encoding first- and second-level neighbourhoods'], fontsize='large')

#plt.show()

plot_path = './benchmark_plot.jpg'
plt.savefig(plot_path, dpi=300)
plt.close()
```
